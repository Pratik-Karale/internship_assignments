# Assignment No 3: Data Preprocessing for Linear Regression
This notebook performs full preprocessing on a car price dataset to prepare it for a linear regression model.

```python
import numpy as np
import pandas as pd
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import LabelEncoder, OneHotEncoder, StandardScaler
from sklearn.compose import ColumnTransformer
from sklearn.model_selection import train_test_split
```

```python
dataset = pd.read_csv('car_data.csv')
X = dataset.iloc[:, [0, 1, 3, 4]].values
y = dataset.iloc[:, 2].values
```

```python
imputer = SimpleImputer(missing_values=np.nan, strategy='mean')
X[:, 1:3] = imputer.fit_transform(X[:, 1:3])
```

```python
ct = ColumnTransformer(transformers=[('encoder', OneHotEncoder(), [0, 3])], remainder='passthrough')
X = np.array(ct.fit_transform(X))
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)
```

```python
sc = StandardScaler()
X_train[:, -2:] = sc.fit_transform(X_train[:, -2:])
X_test[:, -2:] = sc.transform(X_test[:, -2:])
```
